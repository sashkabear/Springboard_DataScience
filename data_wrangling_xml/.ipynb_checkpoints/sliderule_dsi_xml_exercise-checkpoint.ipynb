## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries
for child in document_tree.getroot():
    print(child.find('name').text)
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    capitals_string = ''
    print('* ' + element.find('name').text + ':' + capitals_string)
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print(capitals_string[:-2])
```

## XML exercise

Using data in 'data/mondial_database.xml', the examples above, and refering to https://docs.python.org/2.7/library/xml.etree.elementtree.html, find

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)
4. name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
document = ET.parse( './data/mondial_database.xml' )
```

#### 1. 10 countries with the lowest infant mortality rates

```python
#initiate empty lists to store country names and corresponding infant mortality
countries =[]
inf_mortality =[]
```

```python
#not all aountries contain the infant_mortality element, so need a conditional
#if there is no element put in a NaN
for element in document.getiterator('country'):
    countries.append(element.find('name').text)
    try:
        inf_mortality.append(element.find('infant_mortality').text)
    except:
        inf_mortality.append('NaN')
   
```

```python
import pandas as pd
```

```python
#Make a DataFrame from two lists
df1 = pd.DataFrame({'country' : countries, 'infant_mortality' : inf_mortality})
df1
```

```python
#Before sorting, need to convert items in infant_mortality column from string to float
df1.infant_mortality = [float(i) for i in df1.infant_mortality]
```

```python
#Sort df1 to display lowest infant_mortality values
df1.sort('infant_mortality').head(10)
```

#### 2. 10 cities with the largest population

```python
#Staer emptry listes for cities, year, and population
City = []
Year =[]
Pop = []
```

```python
#Pull out information for city names, their populations, and years of population data

#Iterate over each country
for element in document.getiterator('country'):
    #Iterate over each city in a given country
    for city in element.getiterator('city'):
        #Save name of city
        city_name = city.find('name').text
        #Iterate over each population entry in a given city
        for population in city.getiterator('population'):
            #Populate City list with text of city name attribute (saved above in city_name)
            City.append(city_name)
            #Populate Year list with the Year attribute of population
            Year.append(population.get('year'))
            #Populate Pop list with text of population
            Pop.append(population.text)
            
            
    
        
```

```python
#Build a DataFrame from extracted data
df2 = pd.DataFrame({'city' : City, 'year' : Year, 'population' : Pop})
df2
```

```python
#Convert population and year columns to integers
df2.population = [int(i) for i in df2.population]
df2.year = [int(i) for i in df2.year]
```

```python
#Need to keep only latest population values
df2_clean = df2.drop_duplicates(subset = 'city', keep = 'last')
df2_clean
```

```python
#Sort data frame by population column, in reverse order: from largest to smallest
df2_clean.sort('population', ascending = False).set_index('city').head(10)
```

#### 3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
# Start empty lists for country, year and populations
Country = []
Year2 = []
Pop2 = []
```

```python
#Pull out information for country names, their populations, and years of population data
#Do not include population values of cities and provinces

#Iterate over each country
for element in document.getiterator('country'):
    country_name = element.find('name').text
    #Iterate over each population entry in a given country
    for population in element.getiterator():
        if population.tag == 'population':
            #Populate Country list
            Country.append(country_name)
            #Populate Year list with the Year attribute of population
            Year2.append(population.get('year'))
            #Populate Pop list with text of population
            Pop2.append(population.text)
        # Stop once City of Province elements are encountered - only want country population data
        elif population.tag =='city' or population.tag == 'province':
            break

            
```

```python
#Build a DataFrame from extracted data
df3 = pd.DataFrame({'country' : Country, 'year' : Year2, 'population' : Pop2})
```

```python
#Convert population and year columns to integers
df3.population = [int(i) for i in df3.population]
df3.year = [int(i) for i in df3.year]
```

```python
#Need to keep only latest population values
df3_clean = df3.drop_duplicates(subset = 'country', keep = 'last')
df3_clean
```

```python
# Now pull out data on country name, ethnic group and percentage
Country2 = []
Ethnicity = []
Percent =[]
```

```python
#Iterate over each country
for element in document.getiterator('country'):
    country_name = element.find('name').text
    #Iterate over each ethnicgroup entry in a given country
    for eth in element.getiterator('ethnicgroup'):
        #Populate Country list
        Country2.append(country_name)
        #Populate Percent list with percentage attribute of ethnicgroup
        Percent.append(eth.get('percentage'))
        #Populate Ethnicity list with text of ethnicgroup
        Ethnicity.append(eth.text)
```

```python
#Build a DataFrame from extracted data
df3_b = pd.DataFrame({'country' : Country2, 'ethnicity' : Ethnicity, 'percent' : Percent})
```

```python
#Convert percent columns to float
df3_b.percent = [float(i) for i in df3_b.percent]
```

```python
#Merge two data frames to fill out total country population next to ethnic percentage value
comb = df3_b.merge(df3_clean, how = 'inner')
comb.head(20)
```

```python
#create a column for population of a given ethnicity based on percentage and total country population
comb['eth_pop'] = comb.percent/100*comb.population
```

```python
#drop unnecessary columns, goup by ethnicity, add the population totals, and sort
comb.drop(comb.columns[[0, 2, 3, 4]], axis=1).groupby('ethnicity').sum().sort('eth_pop', ascending = False).head(10)
```
